Now we create the image to grayscale transformer and the HOG transformers as custom transformer functions.

Image to grayscale transformer

```python
class imgtograyTransformer(BaseEstimator,TransformerMixin):
     def __init__(self):
        pass
     def fit(self,X,y=None):
        return self

     def transform(self,X,y=None):
        return np.array([skimage.color.rgb2gray(x) for x in X])
```

```python
rgbobje = imgtograyTransformer()
fds = rgbobje.fit_transform(x_train)
```

Hog transformer

```python
# hog transform
img_re = skimage.transform.rescale(img, (1/3, 1/3, 1))
gray = skimage.color.rgb2gray(img_re)
feature_vector, hog_img = skimage.feature.hog(gray, orientations=9, pixels_per_cell=(14, 14), cells_per_block=(2,2), visualize=True)
```

```python
class hogTransformer(BaseEstimator,TransformerMixin):
     def __init__(self, orientations=9, pixels_per_cell=(8,8), cells_per_block=(3,3)):
        self.orientations = orientations
        self.pixels_per_cell = pixels_per_cell
        self.cells_per_block = cells_per_block

     def fit(self,X,y=None):
        return self

     def transform(self,X,y=None):
        def local_hog(img):
            hog_features = skimage.feature.hog(img, orientations=self.orientations, pixels_per_cell=self.pixels_per_cell, cells_per_block= self.cells_per_block)
            return hog_features
        hfeatures = np.array([local_hog(x) for x in X])
        return hfeatures
```

SGD , SVM, logistic regression, knn neighbors

Here we will test SGDClassifier, SVM, logistic regression and knn neighbors. These are amongst the best models for image data. The others are not good because they try to detect a logic in the data whereas with image data it is usually a spatial pattern that the model must detect.

```python
model_sgd = SGDClassifier(loss='hinge', learning_rate='adaptive', early_stopping=True, eta0=0.1)
```

```python
grayify = imgtograyTransformer()
hogify = hogTransformer()
scalify = StandardScaler()
```

```python
# Convert into grayscale
x_train_gray = grayify.fit_transform(x_train)
```

```python
# Extract the features
x_train_hog = hogify.fit_transform(x_train_gray)
```

```python
# Normalize
x_train_scale = scalify.fit_transform(x_train_hog)
```

```python
# Machine learning
model_sgd.fit(x_train_scale, y_train)
```
